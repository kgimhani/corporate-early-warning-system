```python
import pandas as pd
import numpy as np
np.random.seed(42)

n = 350

companies = [
    "AlphaCorp","BetaInc","GammaTech","DeltaFin","EpsilonMfg",
    "ZetaRetail","EtaEnergy","ThetaMedia","IotaBank","KappaPharm",
    "LambdaAuto","MuConsult","NuLogistics","XiTelecom","OmicronReal",
    "PiConstruct","RhoInsure","SigmaFood","TauApparel","UpsilonMine",
    "PhiAirline","ChiHotel","PsiDefense","OmegaUtil","AlphaTwo",
    "BetaTwo","GammaTwo","DeltaTwo","EpsilonTwo","ZetaTwo"
]

sectors = ["Technology","Finance","Retail","Manufacturing",
           "Energy","Healthcare","Real Estate","Transportation"]

rows = []

for i in range(n):
    company = companies[i % len(companies)]
    sector = sectors[i % len(sectors)]
    year = np.random.choice([2018,2019,2020,2021,2022])

    # latent risk (moderate spread)
    r = np.clip(np.random.normal(0.5, 0.2), 0, 1)

    # IMPORTANT: add randomness to target (not deterministic)
    distressed = np.random.binomial(1, 0.3 + 0.4*(r - 0.5))

    noise = lambda s: np.random.normal(0, s)

    current_ratio   = np.clip(1.8 - 0.8*r + noise(0.3), 0.3, 4)
    debt_to_equity  = np.clip(1.5 + 2.0*r + noise(0.5), 0.1, 6)
    roa             = np.clip(0.08 - 0.12*r + noise(0.05), -0.3, 0.3)
    roe             = np.clip(0.12 - 0.20*r + noise(0.07), -0.5, 0.6)
    gross_margin    = np.clip(0.35 - 0.15*r + noise(0.08), 0.05, 0.8)
    operating_margin= np.clip(0.15 - 0.18*r + noise(0.08), -0.3, 0.5)
    interest_cov    = np.clip(7 - 4*r + noise(1.5), 0.5, 20)
    asset_turnover  = np.clip(1.0 - 0.3*r + noise(0.2), 0.2, 2.5)
    revenue_growth  = np.clip(0.1 - 0.15*r + noise(0.1), -0.4, 0.6)
    cash_ratio      = np.clip(0.4 - 0.2*r + noise(0.1), 0.01, 1)
    altman_z        = np.clip(3.0 - 1.8*r + noise(0.4), 0.2, 6)
    mda_sent        = np.clip(0.2 - 0.4*r + noise(0.2), -1, 1)
    news_sent       = np.clip(0.2 - 0.35*r + noise(0.2), -1, 1)
    days_payable    = np.clip(40 + 30*r + noise(10), 10, 150)

    rows.append({
        "company":company,"sector":sector,"year":year,
        "current_ratio":round(current_ratio,2),
        "debt_to_equity":round(debt_to_equity,2),
        "roa":round(roa,4),"roe":round(roe,4),
        "gross_margin":round(gross_margin,4),
        "operating_margin":round(operating_margin,4),
        "interest_coverage":round(interest_cov,2),
        "asset_turnover":round(asset_turnover,2),
        "revenue_growth":round(revenue_growth,4),
        "cash_ratio":round(cash_ratio,4),
        "altman_z_score":round(altman_z,2),
        "mda_sentiment":round(mda_sent,3),
        "news_sentiment":round(news_sent,3),
        "days_payable_outstanding":round(days_payable,1),
        "distressed":distressed
    })

df = pd.DataFrame(rows)
print("Dataset created successfully")
print("Shape:", df.shape)

print("\nFirst 5 rows:")
print(df.head())

print("\nDistressed distribution:")
print(df["distressed"].value_counts())

print("\nSummary stats:")
print(df.describe().round(2))
```
